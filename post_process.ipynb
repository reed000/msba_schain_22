```python
from scipy import optimize
import numpy as np
from matplotlib import pyplot as plt

def f1(x):
    return 0.6*(2 - x)**2

def f2(x):
    return np.sin(x) + x + x * np.sin(x)

### sum of f1 and f2
def f3(x):
    return f1(x[0]) + f2(x[1])
```

```python
help(optimize.minimize)
```

```python
optimize.minimize(f3, [0,0], method="Nelder-Mead")
```

```python
rranges = (slice(-10, 10, 0.1), slice(-10, 10, 0.1))
resbrute = optimize.brute(f3, rranges, full_output=True, 
                              finish=optimize.fmin)
resbrute
```
